```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for 15 persons
np.random.seed(42)  # Set seed for reproducibility

# Randomly generate 15 ages between 20 and 60
ages = np.random.randint(20, 60, size=15)

# Randomly generate 15 income values between 30,000 and 150,000
incomes = np.random.randint(30000, 150000, size=15)

# Create a DataFrame with this random data
df = pd.DataFrame({
    'Age': ages,
    'Income($)': incomes
})

# Display the DataFrame
print("Generated Data:")
print(df)

# Preprocessing: Normalize the Age and Income columns using MinMaxScaler
scaler = MinMaxScaler()
df[['Age', 'Income($)']] = scaler.fit_transform(df[['Age', 'Income($)']])

# Apply K-Means clustering with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[['Age', 'Income($)']])

# Print the cluster centers
print("\nCluster centers:")
print(kmeans.cluster_centers_)

# Plotting the clusters
plt.scatter(df['Age'], df['Income($)'], c=df['Cluster'], cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', marker='X')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.title('K-Means Clustering (K=3) - Random Data')
plt.show()
```
